# Machine Learning
### Machine Learning (Pembelajaran Mesin) adalah cabang dari kecerdasan buatan (Artificial Intelligence, AI) yang berfokus pada pengembangan algoritma dan model yang memungkinkan komputer untuk "belajar" dari data tanpa harus diprogram secara eksplisit untuk setiap tugas. 
- Dalam machine learning, sistem mengidentifikasi pola dan struktur dalam data, dan kemudian menggunakan pola-pola tersebut untuk membuat prediksi atau keputusan.
- Inti dari machine learning adalah menggunakan data untuk mengajarkan komputer cara membuat keputusan atau prediksi yang semakin baik seiring waktu tanpa intervensi manusia yang terus menerus

# Komponen Utama Machine Learning:
- Data: Sumber utama pembelajaran. Data bisa berupa teks, gambar, angka, atau data lain yang dapat digunakan untuk memodelkan suatu masalah (relevan, Lengkap, Bersih, Representatif, Jumlah Cukup).
- Fitur (Features): Atribut atau variabel dalam data yang digunakan untuk membantu algoritma mempelajari pola. Misalnya, dalam data penjualan, fitur bisa berupa harga, lokasi, dan waktu pembelian.
- Model: Model adalah represetasi matematis hasil dari proses pelatihan algoritma dengan data tertentu (Kompleksitas, Interpretabilitas).
- Algoritma: Merupakan prosedur matematika atau langkah-langkah komputasi yang digunakan untuk membangun dan melatih model machine learning (Jenis Masalah, Ukuran Data, Komputasi & Waktu). 
- Pelatihan (Training): Proses di mana model machine learning belajar dari data. Model dibangun dengan mempelajari hubungan ataupun pola data (Data Pelatihan, Optimasi, Hasil).
- Evaluasi & Validasi: Tahap mengukur performa model dengan pola data baru yang belum pernah digunakan dalam tahap pelatihan model. Tahap ini berguna untuk menilai kinerja dan performa akhir dari model machine learning (Overfitting, Underfitting)

# Pengumpulan Data
- Pengumpulan data adalah tahap awal di mana dataset dikumpulkan dari berbagai sumber agar bisa digunakan dalam pelatihan model. Sumber data adalah aset utama untuk machine learning, karena kualitas data merupakan faktor penentu kinerja model.
1. Data internal organisasi (perusahaan)
    - Data internal organisasi (perusahaan) adalah yang dihasilkan, dikumpulkan, atau dikelola secara langsung oleh organisasi (perusahaan) melalui sistem internalnya.
    - Seperti sistem penjualan (data transaksi pelanggan), sistem CRM (data profil pelanggan, histori pembelian, interaksi layanan, dll).
2. Data eksternal
    - Merupakan data yang disediakan oleh sumber luar organisasi, bisa dibeli, diperoleh via mitra, atau sumber data terbuka (open data).
    - Contoh sumber data seperti dataset publik terbuka (data demografis, lingkungan, ekonomi, cuaca), data lingkungan/regulasi (standar industri, peraturan pemerintah), open data portals: (data pemerintah, World Bank, UN, Kaggle, UCI) dan dataset berlisensi komersial (cranimed data, data pemasaran, data kepuasan pelanggan).


3. Data sensor/IOT
    - Data sensor/IoT adalah data yang dihasilkan oleh perangkat fisik yang terhubung ke jaringan internet atau jaringan lokal.
    - Biasanya data ini berupa stream waktu nyata (real-time) yang mencatat nilai-nilai relevan seperti temperatur, tekanan, kelembapan, getaran, lokalisasi, serta metadata perangkat.
    - Contoh sumber data ini, seperti data sensor industry (suhu mesin, getaran, tekanan, aliran), data sensor lingkungan (suhu ruangan, kelembapan, kualitas udara, CO2), data kamera vision-based sensors (frame-level data dengan informasi metadata), data sistem kendaraan terhubung (telemetri kendaraan, GPS, akselerasi).
4. Data crowdsourcing
    - Data crowdsourcing adalah data yang dikumpulkan dari kontributor publik, karyawan, atau komunitas melalui partisipasi terbuka maupun tersegmentasi.
    - Contoh umum meliputi anotasi gambar, terjemahan teks, labeling suara, ataupun pengumpulan data sensor buatan pengguna.
    - Contoh sumber data ini adalah data anotasi gambar/marker objek di dataset visual (labeling bounding box, polygon), data terjemahan bahasa, perbaikan teks, evaluasi kualitas QA, data lokasi/kegiatan dari kontribusi pengguna (crowdsourcing geospasional, laporan komunitas), data ulasan yang dikumpulkan melalui komunitas pengguna.


# Data Cleaning & Data Preprocessing
- Data cleaning dan data preprocessing adalah langkah-langkah penting dalam alur kerja data science dan machine learning. Mereka memastikan data yang digunakan untuk analisis dan pelatihan model berkualitas, konsisten, dan relevan.
- Data cleaning adalah proses mengidentifikasi dan memperbaiki kesalahan, ketidakkonsistenan, serta kekurangan dalam data. Tujuannya adalah meningkatkan kualitas data sehingga analisis atau model yang dibangun lebih akurat.

1. Data tidak relevan
    - Merupakan kolom/fitur yang tidak informatif atau berdampak rendah terhadap tujuan analisis/model.
    - Penangananya dengan menghapus kolom tersebut
2. Data duplikat (duplicate)
    - Merupakan kolom/fitur yang duplikasi/ganda.
    - Penangananya dengan menghapus kolom tersebut
3. Data missing values (nilai hilang)
    - Merupakan kolom/fitur yang kehilangan nilai
    - Penanganannya dengan menghapus baris/kolom yang missing value, Mengimputasi nilai hilang dengan nilai mean/median/modus.
4. Data outlier
    - Merupakan kolom/fitur yang memiliki nilai yang tidak lazim
    - Penanganannya dengan cara menghapus baris/ data tersebut, menggunakan strategi statistik (z-score, IQR) atau domain knowledge.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# Data contoh
df = pd.DataFrame({
    'Id': ['k01','k02','k03','k04','k05'],
    'Nama': ['Deni','Ani','Jeni','Alam','Lala'],
    'Alamat': ['Jakarta','BSD','Pamulang','Tangerang','Pamulang'],
    'Pembelian': [13.0, 45.0, 2000.0, 11.0, 517.25]
})

# 1) Pastikan kolom numeric (jaga NaN jika ada)
df['Pembelian'] = pd.to_numeric(df['Pembelian'], errors='coerce')

# 2) Hitung Z-score untuk kolom Pembelian
z = stats.zscore(df['Pembelian'], nan_policy='omit')

# 3) Tentukan ambang outlier
ambang = 2.0  # bisa diubah jadi 3.0

# 4) Deteksi outlier
outlier_mask = np.abs(z) >= ambang

# 5) Hapus baris yang outlier
df_clean = df.loc[~outlier_mask].copy()

print("Data asli:")
print(df)
print("\nData setelah menghapus outlier (Z-score):")
print(df_clean)
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats

# 1) Pastikan kolom numeric
df1['Pembelian'] = pd.to_numeric(df1['Pembelian'], errors='coerce')

# 2) Hitung Z-score
z = stats.zscore(df1['Pembelian'], nan_policy='omit')

# 3) Ambang outlier
ambang = 2.0

# 4) Mask outlier
outlier_mask = np.abs(z) >= ambang

# 5) Hapus baris outlier
df_clean = df1.loc[~outlier_mask].copy()

print("Data asli:")
print(df1)
print("\nMask outlier:")
print(outlier_mask)
print("\nData setelah menghapus outlier (Z-score):")
print(df_clean)
```
